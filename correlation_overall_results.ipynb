```python
acc_data = pd.read_csv('sentimet140_acc_analysis.csv')
zz = acc_data.drop(index=0)
```

```python
zz
```

```python
data_list = ['sentimet140','conversational','amazon','semeval','newsheadline','fused']
```

```python
print(str(data_list[0])+'I am fine')
```

```python
for datset in data_list:
    xx=pd.read_csv(str(datset)+'_acc_analysis.csv')
    zz = pd.concat([zz,xx],axis=0)
```

```python
zz
```

```python
ax = zz.plot(x="Dataset", y=["Log_reg_meta","GBoost_meta","VADER","TextBlob","Flair","BERT"],kind="bar",figsize=(16.5,9.5))
ax.legend(["Log_reg_meta","GBoost_meta","VADER","TextBlob","Flair","BERT"])
ax.set_xlabel('Dataset', fontsize=14)  # Change label and font size
ax.set_ylabel('Accuracy', fontsize=14)  # Change label and font size
ax.set_title('Accuracy Scores : Ensemble Learning (Log Regressor , Grad Boost)', fontsize=16)
y_ticks = [(i*.05) for i in range(0,20)]
ax.set_yticks(y_ticks)
ax.yaxis.grid(True)
# Increase font size of tick labels
ax.plot(zz['Dataset'], zz['Log_reg_meta'], color='blue', marker='o', label='Log_reg_meta')
ax.plot(zz['Dataset'], zz['GBoost_meta'], color='orange', marker='o', label='GBoost_meta')
ax.plot(zz['Dataset'], zz['Flair'], color='purple', marker='o', label='Flair')
ax.plot(zz['Dataset'], zz['BERT'], color='brown', marker='o', label='BERT')
#ax.plot(zz['Dataset'], zz['TextBlob'], color='red', marker='o', label='BERT')
#ax.plot(zz['Dataset'], zz['VADER'], color='green', marker='o', label='BERT')
ax.tick_params(rotation = 25,axis='both', labelsize=12)
ax.legend(loc = 'lower center',fontsize=14)
plot.grid(True)
```
